**Machine Learning in Simple Terms**

Machine learning is a way that computers can get better at doing things on their own, without being explicitly programmed to do so. It's like how you get better at riding a bike or playing a game over time, but with computers!

Here's a simple analogy to help you understand:

**Imagine a Child Learning to Recognize Animals**

You show a child pictures of different animals, like cats, dogs, and birds. At first, the child might not be able to tell them apart. But as they see more pictures and learn to associate certain features (like whiskers or a long tail) with each animal, they start to recognize them more easily.

**Machine Learning Works in a Similar Way**

A computer program, called a **model**, looks at a large collection of data (like pictures of animals). It tries to identify patterns and relationships between different features (like whiskers and tail length) and the labels (like "cat" or "dog"). As the model sees more data, it gets better at recognizing these patterns and making predictions.

**Types of Machine Learning**

There are three main types of machine learning:

1. **Supervised Learning**: The model is trained on labeled data (like pictures of animals with labels) and learns to make predictions based on that.
2. **Unsupervised Learning**: The model is given unlabeled data and tries to find patterns or groupings on its own.
3. **Reinforcement Learning**: The model learns by interacting with an environment and receiving feedback (like rewards or penalties) for its actions.

**Key Benefits**

Machine learning has many applications, including:

* **Improved accuracy**: Machines can make predictions and decisions more accurately than humans in some cases.
* **Automated tasks**: Machines can perform tasks that would take humans a long time to do, like data analysis or image recognition.
* **Personalization**: Machines can learn to tailor their responses to individual users, like recommending products based on their preferences.

#EXAMPLE OF USING LCEL:


```python
from langchain_core.runnables import RunnableLambda

def to_upper(text: str) -> str:
    return text.upper()

def reverse_text(text: str) -> str:
    return text[::-1]

def add_length_info(text: str) -> str:
    return f"{text} (length: {len(text)})"

chain = RunnableLambda(to_upper) | RunnableLambda(reverse_text) | RunnableLambda(add_length_info)

print(chain.invoke("langchain pipes"))
```

```python
from google.colab import userdata
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain.memory import ChatMessageHistory

# Load API Key
NEBIUS_KEY = userdata.get("NEBIUS_API_KEY")
if not NEBIUS_KEY:
    raise RuntimeError("Missing NEBIUS_API_KEY secret in Colab. Add it in 'Secrets'.")

os.environ["OPENAI_API_KEY"] = NEBIUS_KEY
os.environ["NEBIUS_API_KEY"] = NEBIUS_KEY
NEBIUS_BASE = "https://api.studio.nebius.com/v1"

# LLM
llm = ChatOpenAI(
    model="meta-llama/Meta-Llama-3.1-8B-Instruct",
    base_url=NEBIUS_BASE,
    api_key=NEBIUS_KEY,
    temperature=0.6,
    max_tokens=512,
)

# Prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful, concise assistant. Keep replies under 6 lines."),
    MessagesPlaceholder(variable_name="history"),
    ("human", "{input}")
])

# Chain
chain = prompt | llm | StrOutputParser()

# Single shared history
history = ChatMessageHistory()

# Wrap with history
with_history = RunnableWithMessageHistory(
    chain,
    lambda _session_id: history,
    history_messages_key="history",
    input_messages_key="input",
)

def chat_loop():
    print("(commands: /clear, /quit)")
    while True:
        try:
            user = input("You: ").strip()
        except EOFError:
            break
        if not user:
            continue
        if user == "/quit":
            break
        if user == "/clear":
            # Recreate or clear the history
            try:
                history.clear()
            except Exception:
                history.messages = []
            print("Bot: memory cleared.")
            continue

        # pass a constant session_id
        resp = with_history.invoke(
            {"input": user},
            config={"configurable": {"session_id": "single"}}
        )
        print("Bot:", resp)

# Run
chat_loop()
```

```python


from google.colab import userdata
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain.memory import ChatMessageHistory

NEBIUS_KEY = userdata.get("NEBIUS_API_KEY")
if not NEBIUS_KEY:
    raise RuntimeError("Missing NEBIUS_API_KEY secret in Colab. Add it in 'Secrets'.")

os.environ["OPENAI_API_KEY"] = NEBIUS_KEY
os.environ["NEBIUS_API_KEY"] = NEBIUS_KEY

NEBIUS_BASE = "https://api.studio.nebius.com/v1"

#setup
llm = ChatOpenAI(
    model="meta-llama/Meta-Llama-3.1-8B-Instruct",
    base_url=NEBIUS_BASE,
    api_key=NEBIUS_KEY,
    temperature=0.6,
    max_tokens=512,
)

#generic ah prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful, concise assistant. Keep replies under 6 lines."),
    MessagesPlaceholder(variable_name="history"),
    ("human", "{input}")
])

#langCHAIN lol
chain = prompt | llm | StrOutputParser()

#session storag
SESSION_STORE: dict[str, ChatMessageHistory] = {}

def get_session_history(session_id: str) -> ChatMessageHistory:
    if session_id not in SESSION_STORE:
        SESSION_STORE[session_id] = ChatMessageHistory()
    return SESSION_STORE[session_id]

#memory pattern
with_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    history_messages_key="history",
    input_messages_key="input",
)

def chat_loop(initial_session="student-1"):
    session_id = initial_session
    print("(commands: /new <id>, /clear, /quit)")
    print(f"Session: {session_id}")
    while True:
        try:
            user = input("You: ").strip()
        except EOFError:
            break
        if not user:
            continue
        if user == "/quit":
            break
        if user == "/clear":
            SESSION_STORE[session_id] = ChatMessageHistory()
            print("Bot: memory cleared.")
            continue
        if user.startswith("/new"):
            parts = user.split(maxsplit=1)
            session_id = parts[1] if len(parts) > 1 else f"user-{len(SESSION_STORE)+1}"
            print(f"Switched to session: {session_id}")
            continue

        resp = with_history.invoke(
            {"input": user},
            config={"configurable": {"session_id": session_id}},
        )
        print("Bot:", resp)

# run it
chat_loop()
```
