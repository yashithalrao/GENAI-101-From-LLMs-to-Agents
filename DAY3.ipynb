USING NEIBUS AI STUDIO DIRECTLY

```python
import os
from openai import OpenAI

client = OpenAI(
    base_url="https://api.studio.nebius.com/v1/",
    api_key="yyour apikey ")

response = client.chat.completions.create(
    model="meta-llama/Meta-Llama-3.1-8B-Instruct",
    max_tokens=512,
    temperature=0.6,
    top_p=0.9,
    extra_body={
        "top_k": 50
    },
    messages=[{
        "role": "user",
        "content": "Give good restaurants near PES University."
    }]
)

print(response.to_json())

message = response.choices[0].message.content.strip()

print(message)
```

## SINGLE TURNING

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import ChatOpenAI

load_dotenv()
# Assign them to env vars so LangChain’s validation passes
os.environ["NEBIUS_API_KEY"] = NEBIUS_KEY
os.environ["OPENAI_API_KEY"] = OPENAI_KEY or NEBIUS_KEY  # LangChain needs this

NEBIUS_BASE = "https://api.studio.nebius.com/v1"




llm = ChatOpenAI(
    model="meta-llama/Meta-Llama-3.1-8B-Instruct",
    temperature=0.6,
    max_tokens=512,
    openai_api_base=NEBIUS_BASE,
    openai_api_key=NEBIUS_KEY
)
```

```python
prompt = "What is the capital of France?"
print(llm.invoke(prompt).content)
```

```python
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory

conversation = ConversationChain(llm=llm, memory=ConversationBufferMemory())

print(conversation.predict(input="What is the capital of France?"))
print(conversation.predict(input="What is its population?"))
print(conversation.predict(input="What are its famous landmarks?"))
```

```python

conv = ConversationChain(llm=llm, memory=ConversationBufferMemory())
print(conv.run("Give good restaurants near PES University."))
```

```python
from langchain.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain the concept of {topic} in simple terms."
)

#send ips
filled_prompt = template.format(topic="machine learning")

# model
response = llm.invoke(filled_prompt)
print(response.content)
```
