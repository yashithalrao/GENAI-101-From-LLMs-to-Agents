```python


import numpy as np
from sentence_transformers import SentenceTransformer

# Load a semantic embedding model
embedder = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
# Mini knowledge base
docs = [
    "Students must maintain 75% attendance in each course.",
    "Library late fee is ₹5 per day per book.",
    "WiFi password rotates on the first Monday of every month.",
    "Medical exemptions must be submitted within 7 days of return."
]
```

```python
# Create embeddings for docs
doc_vecs = embedder.encode(docs, normalize_embeddings=True)
doc_vecs = np.array(doc_vecs, dtype="float32")
```

```python
# Cosine similarity search
def retrieve(query, k=2):
    q_vec = embedder.encode([query], normalize_embeddings=True)[0]
    sims = doc_vecs @ q_vec  # cosine similarity (normalized vectors)
    idxs = np.argsort(-sims)[:k]
    return [(docs[i], float(sims[i])) for i in idxs]
```

```python
# Try it
queries = [
    "Penalty for overdue books?",
    "When does the WiFi password change?",
    "Attendance rule to write exam?",
    "whos the chairman?"
]

for q in queries:
    print(f"\nQ: {q}")
    results = retrieve(q, k=1)
    for txt, score in results:
        print(f"Retrieved: {txt}\nSimilarity: {score:.3f}")
```
